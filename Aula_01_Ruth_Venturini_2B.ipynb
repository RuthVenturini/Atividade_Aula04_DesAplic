Exemplo 1 - Criando um novo mapa

```python
import folium
mapa = folium.Map (location = [-25.433390,-49.265342],
               tiles = 'Cartodb Positron',
               zoom_start = 11,
               attr = 'Colaboradores do OpenStreetMap e Alunos de CartoUFPR/ Dados IPPUC'
              )
mapa
```

Três pontos turísticos de Curitiba

```python
import folium

tooltip = 'Clique aqui!'

folium.Marker([-25.384618,-49.276169], popup='<b>Curitba</b><br><img src="https://arquitetura.vivadecora.com.br/wp-content/uploads/2022/09/%C3%93pera-de-Arame-vista-de-cima-foto-curtacuritiba.com_.br_.jpeg" style="width:128px;height:128px;"><br><a href="https://www.instagram.com/operadearameoficial/">Instagram Ópera de Arame</a> ',
              tooltip=tooltip,
             icon = folium.Icon(icon='star', color = 'gray'),
             ).add_to(mapa)

folium.Marker([-25.443144,-49.238181], popup='<b>Curitba</b><br><img src="https://mid.curitiba.pr.gov.br/2021/00310996.jpg" style="width:128px;height:128px;"><br><a href="https://www.curitiba.pr.gov.br/conteudo/jardim-botanico-municipal-de-curitiba/287">Site da Prefeitura de Curitiba - Página do Jardim Botânico</a> ',
              tooltip=tooltip,
             icon = folium.Icon(icon='star', color = 'green'),
             ).add_to(mapa)

folium.Marker([-25.434833,-49.256923], popup='<b>Curitba</b><br><img src="https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRcEGfZguTCjsw6MquvgtV0HKtKFikZ2Xxsg_uVFHpvV4-Jv2WLIpaoiqWihLP38kp98vo&usqp=CAU" style="width:128px;height:128px;"><br><a href="https://www.mercadomunicipaldecuritiba.com.br/">Site do Mercado Municipal de Curitiba</a> ',
              tooltip=tooltip,
             icon = folium.Icon(icon='star', color = 'red'),
             ).add_to(mapa)

mapa.save('/content/drive/My Drive/2025 - 1 - Desenvolvimento de aplicações Geoespaciais/Aula 01/meumapa.html ')
mapa
```
